```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_openml
from sklearn.manifold import TSNE
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
plt.style.use('ggplot')
```

```python
mnist = fetch_openml('mnist_784', version=1)
X = mnist.data
y = mnist.target
```

```python
X
```

```python
X1 = X[:40000].copy()
```

```python
y
```

```python
y1 = y[:40000].copy()
```

```python
X1 = StandardScaler().fit_transform(X1)
```

```python
X1.shape
```

```python
X_Embedded = TSNE(n_components = 2, perplexity = 50, early_exaggeration = 4.5, learning_rate = 960, random_state = 0, ).fit_transform(X1)
```

```python
X_Embedded.shape
```

```python
y1 = y1.astype(float)
y1
```

```python
plt.figure(figsize=(18,12))
plt.scatter(x = X_Embedded[:,0], y = X_Embedded[:,1], c = y1, cmap = 'tab10');
```

```python
pca = PCA(n_components=2)
pca_X = pca.fit_transform(X1)
```

```python
plt.figure(figsize=(18,12))
plt.scatter(x = pca_X[:,0], y = pca_X[:,1], c = y1, cmap = 'tab10');
```
